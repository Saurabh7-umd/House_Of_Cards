<b>Rules of Game:</b><br>
<b>1) Use only cards with numbers 2 to 10, totaling 36 cards.</b><br>
<b>2) Disregard the card suits, they won't affect the game.</b><br>
<b>3) The goal is to get rid of all your cards.</b><br>

<b>Points System:</b><br>
Each card is worth the number displayed on it (e.g., 6 of hearts is worth 6 points).

<b>Game Objective:</b><br>
The winner is the player with the lowest score.<br>
The game ends when one player runs out of cards in a round or after 50 rounds.

<b>Playing a Round:</b><br>
Each player starts with 2 random cards.<br>
Calculate the total points of these 2 cards.

<b>Winning a Round:</b><br>
The player with the highest total points wins the round.<br>
The loser takes all 4 cards played in the round.

<b>Tie Breaker:</b><br>
If there's a tie, players keep drawing pairs of cards until the tie is broken.

<b>Ending the Game:</b><br>
The game ends when a player can't complete a round or after 50 rounds.

<b>Determining the Winner:</b><br>
The player with the lowest total points wins the game.


<b>Creation of Deck:</b><br>
To represent a standard deck of playing cards, you can use a list of sublist, where each sublist represents a card. Each sublist contains two elements: the rank and the suit of the card.

```python
# Import necessary libraries
import random
import statistics

# Define suits and values for the deck.
suits = ["Spades", "Hearts", "Diamonds", "Clubs"]
values = [2, 3, 4, 5, 6, 7, 8, 9, 10]

# Initialize an empty deck
deck=[]
# Create the deck by combining suits and values
for suit in suits:
    for value in values:
        deck.append([suit,value])
print('Cards in deck: ',deck)
```
